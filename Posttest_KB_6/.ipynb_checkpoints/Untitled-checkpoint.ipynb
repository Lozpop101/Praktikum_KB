```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("prepro_ford.csv")
df.head(5)
```

```python
df.isna().sum()
```

# Clustering

## K-Means Clustering

```python
point = df[['price','year']]
point = point.values
```

```python
model = KMeans(n_clusters=5)
model.fit(point)
```

```python
y = model.fit_predict(point)
print(y)
```

```python
centroids = model.cluster_centers_

centroid_x = centroids[:,0]
centroid_y = centroids[:,1]

plt.figure(figsize=(10,10))
plt.scatter(point[y==0,0], point[y==0,1], c='lightblue', label='mobil 1', alpha=0.5)
plt.scatter(point[y==1,0], point[y==1,1], c='red', label='mobil 2', alpha=0.5)
plt.scatter(point[y==2,0], point[y==2,1], c='green', label='mobil 3', alpha=0.5)
plt.scatter(point[y==3,0], point[y==3,1], c='purple', label='mobil 4', alpha=0.5)
plt.scatter(point[y==4,0], point[y==4,1], c='blue', label='mobil 5', alpha=0.5)


plt.scatter(centroid_x, centroid_y, marker="D", s=50, label="Centroid", c='black')
plt.xlabel("Harga/Price")
plt.ylabel("Tahun dibuat")
plt.legend()
plt.show()
```

## Hierarchical Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage, complete
```

```python
linkage_hr = linkage(point, method='ward')
complete_hr = complete(point)
```

```python
plt.figure(figsize=(10,7))
dendrogram(linkage_hr)
plt.title("Dendrogram Linkage")
plt.xlabel("Harga")
plt.ylabel("Tahun di Buat")
plt.show()
```
