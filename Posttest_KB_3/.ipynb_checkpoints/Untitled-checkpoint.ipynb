```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Airbnb_Open_Data.csv", low_memory=False)
df.head()
```

```python
df.info()
```

```python
data = df[["review rate number", "Construction year", "minimum nights", "availability 365", "number of reviews", "neighbourhood group"]]
data.head(2)
```

```python
data = data.rename(columns = {'review rate number':'rating', 
                              'Construction year':'tahun_pembuatan', 
                              'minimum nights':'malam_minimal',
                                'availability 365':'Ketersediaan',
                             "number of reviews":"jumlah review",
                             "neighbourhood group":"group_tetangga"})
data.head()
```

```python
data_pie = data['rating'].value_counts().rename_axis('rating').reset_index(name='id')
```

```python
plt.figure(figsize=(10,10))

plt.pie(data_pie.id, labels=data_pie.rating, startangle=90, autopct='%.1f%%')

plt.title('Rating Dalam Review')
plt.show
```

Dalam pie chart diatas dapat diketahui bahwa dari sekian banyak Airbnb yang ada didalam dataset, jumlah yang mendapat rating 5 sebanyak 22.8% sama dengan rating 4 juga sebanyak 22.8%, rating 3 sebanyak 22.7, rating 2 sebanyak 22.6% dan rating 1 sebanyak 9%. 




```python
data_pie = data['group_tetangga'].value_counts().rename_axis('group_tetangga').reset_index(name='id')
```

```python
plt.figure(figsize=(10,10))

plt.pie(data_pie.id, labels=data_pie.group_tetangga, startangle=0, autopct='%.1f%%')

plt.title('Pembagian Lokasi')
plt.show
```

Dalam pie chart diatas dilhat pembagian lokasi Airbnb dalam dataset, di Manhattan 42.7%, di Brooklyn 40.8%, di Queens 12.9%, di Bronx 2.6%, di Staten Island 0.9%. Jadi dalam data tersebut jumlah Airbnb yang paling banyak adalah di daerah Manhattan.

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(data.corr(), center=0, cmap='Purples')
_=ax.set_title('Korelasi antar atribut pada data Airbnb.csv')
```

Gambar diatas menunjukkan bahwa data yang memiliki korelasi paling tinggi adalah data reviews per month dan number of reviews, sedangkan data lain korelasinya kurang 0.2

```python
plt.scatter(data = df, x ='reviews per month', y ='number of reviews')
plt.title("Scatterplot")
plt.xlabel("reviews per month")
plt.ylabel("Total reviews")
plt.show
```
