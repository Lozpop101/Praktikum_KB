```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("prepro_ford.csv")
df.head(10)
```

```python
x = df.drop('engineSize', axis=1)
y = df['engineSize']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 12)
```

```python
X_train = tf.expand_dims(x_train, axis=0)
Y_train = tf.expand_dims(y_train, axis=0)
```

```python
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
print(y_train[0])
```

```python
model = keras.Sequential([
    layers.Dense(units=64, input_shape=[32]),
    layers.Dense(32, activation='relu'),
    layers.Dense(32, kernel_initializer=tf.keras.initializers.he_normal),
    layers.Dense(32, kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    layers.Dense(10, activation='softmax'),
])

model.summary()
```

```python
Optimizer = optimizers.Adam(0.001)
model.compile(loss=keras.losses.categorical_crossentropy, optimizer=Optimizer,
metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=100, validation_data=(x_test, y_test))
```

```python
model.evaluate( x_test, y_test, batch_size=32)
```

```python
result = model.predict(x_test)
print(result)
```

```python
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
```
